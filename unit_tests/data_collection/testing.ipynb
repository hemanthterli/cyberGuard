```python
import cv2
from rapidocr_onnxruntime import RapidOCR

# 1. Initialize the OCR engine
ocr = RapidOCR()
```

```python

# 2. Path to your image
image_path = r"H:\PGAGI\cyberGuard\test_data\cyberGuard_test1.jpeg"

# 3. Perform OCR
# result is a list containing [box, text, score]
result, elapse = ocr(image_path)

# 4. Process and Print Results
if result:
    for text_info in result:
        # text_info[1] contains the extracted text
        print(f"Detected Text: {text_info[1]}")
else:
    print("No text detected.")

# Optional: Visualize the result
# img = cv2.imread(image_path)
# from rapidocr_onnxruntime import vis_res
# vis_res(result, img, "result.jpg")
```
